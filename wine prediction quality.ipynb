```python
#in this dataset target variable (i.e dependent variable )is quality
#so now drop the column from the dataset
x=df.drop('quality',axis=1)
x
```

```python
y=df['quality']
y
```

```python
#now split the dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=60,stratify=y)

```

```python
#define import modules thw modules 

from sklearn.linear_model import LogisticRegression
```

```python
#apply the model
# from sklearn.linear_model import LinearRegression
regr=LogisticRegression()
#fit the model
regr.fit(x_train,y_train)
#predict thhe model
xtrainprediction=regr.predict(x_train)
xtrainprediction
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy=accuracy_score(xtrainprediction,y_train)
accuracy
```
